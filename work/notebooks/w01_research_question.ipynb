## 1. My lane (or freestyle) and why

**Lane 2: Refresh / Content Opportunity Scoring**

As an AI graduate, this lane provides a strong applied machine learning challenge. It requires predicting outcomes from observational data, constructing a robust ranking system evaluated with top-K metrics, and dealing with real-world problems like structured missingness and explainability via reason codes.

## 2. The question: decision, action, cost of a wrong call

- **Decision**: Which visible pages should a content editor prioritize for a refresh review?
- **Action**: The editor reviews the top-ranked pages in the queue, checks the suggested reason codes, and decides whether to update the metadata, expand the content, or monitor it.
- **Cost of wrong call**: Wasted editor hours reviewing pages that have no real problem, or missing high-impact pages where a refresh could yield traffic gains.
- **Why ML**: The factors influencing traffic decline (age, CTR, search volume changes, position drops) interact in complex ways that a simple static rule will either over-flag (too many false positives) or under-flag.

## 3. Quick look at the data (2-3 real numbers)

The data shows a massive backlog of high-impact pages that need review, proving that an algorithmic ranking is required to triage them.

```python
import pandas as pd

df = pd.read_csv('../../data/raw/content_refresh_anonymized.csv')
declining_with_impact = len(df[(df['trend_direction'] == 'down') & (df['impressions_90d'] > 500)])
stale_with_impact = len(df[(df['content_age_days'] > 180) & (df['impressions_90d'] > 500)])

print(f"Total pages in starter dataset: {len(df)}")
print(f"Pages showing declining trends with >500 recent impressions: {declining_with_impact}")
print(f"Stale pages (>180 days old) with >500 recent impressions: {stale_with_impact}")
print("Conclusion: There are nearly 10,000 pages with high visibility that are either stale or declining. A manual review of all 10,000 is impossible, making a ranking model essential.")
```

## 4. Careful words: what I can and can't claim

- **Can claim**: I can claim that certain observable content features (like age or CTR gaps) correlate with traffic decline. I can provide a directional, decision-support score to rank pages for a human reviewer.
- **Cannot claim**: I cannot claim that updating these pages will guarantee a recovery (no causal proof). I cannot claim to predict or reverse-engineer Google's algorithm factors.
